```python
def getIMDB_info(tconst):
    # create an instance of the IMDb class
    ia = IMDb()
    
    # get a movie
    movie = ia.get_movie(tconst.replace("tt",""))
    
    # return language of the movie
    movie_lang=''
    for lang in movie['language']:
        movie_lang=movie_lang+lang
    return movie_lang
```
